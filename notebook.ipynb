### Consumption and disposable income (page 47) with savings (page 82)

Formula for consumption C = c0 + c1*YD 
Formula for saving S = -c0 + (1 - c1)*YD,
where C is consumption, c1 is marginal propensity to consume and YD is dispensible income.

Dispensible income equals the difference between income and taxes.
Formula for saving
YD = Y - T,
where Y is income and T is taxes.


```python
import ipywidgets
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
# Yd is dispensible income
Yd = np.linspace(0, 150, num=2) 
c0 = 50
c1 = 0.5
# the consumption function
def consumption(Yd, c0, c1):
    return c0 + c1*Yd

def saving(Yd, c0, c1):
    return -c0 + (1-c1)*Yd

fig, ax = plt.subplots()

## Draw the initial plot
#surf1 = ax.plot(Yd, consumption(np.ravel(Yd), c0, c1), label="consumption")
#surf2 = ax.plot(Yd, saving(np.ravel(Yd), c0, c1), label="saving")

ax.set_xlim([0, 100])
ax.set_ylim([0, 100])

# set axis labels
ax.set_xlabel("disposable income(Y)")
ax.set_ylabel('consumption and saving')
    

# set legend 
#plt.legend(["consumption", "saving"], loc="upper left")


# update plot and show slides
slider_c0 = ipywidgets.IntSlider(value=10, min=0, max=100, description = "c0", continuous_update = True)
slider_c1 = ipywidgets.FloatSlider(value=0.5, min=0, max=1, description = "c1", continuous_update = True)


def update(c0=10, c1=0.9):
    # clearing and rebuilding the graph with new values for c0 and c1.
    # Not the most efficient method, but it gets the job done.
    ax.clear()

    ax.set_xlim([0, 100])
    ax.set_ylim([0, 100])

    # set axis labels
    ax.set_xlabel("disposable income(Y)")
    ax.set_ylabel('consumption and saving')
    # Draw the initial plot
    ax.plot(Yd, consumption(np.ravel(Yd), c0, c1), label="consumption")
    ax.plot(Yd, saving(np.ravel(Yd), c0, c1), label="saving")
    ax.legend(["consumption", "saving"], loc="upper left")

  
interactive_plot_1 = ipywidgets.interactive(update, c0=slider_c0, c1=slider_c1)
display(interactive_plot_1)
```

### Demand and the determination of equilibrium output (pages 73 - 75)

Assuming that imports and exports are both 0 (or equal), the demand is equal to the sum of consumption, investments and government spending.

Z = C + I + G

We know that C = c0 + c1*(Y - T)

Z = c0 + c1*(Y - T) + I + G

We also know that production equals the sum of consumption, investments and government spending.

Y = C + I + G

Therefore demand equals production

Y = Z

Replacing Z with C + I + G leads to

Y = C + I + G

and

Y = c0 + c1*(Y - T) + I + G

"In equilibrium, production, Y (the left side of the equation), is equal to demand (the right side). Demand in turn depends on income, Y, which is itself equal to production" (Macroeconimics by Olivier Blanchard and David R. Jonhson, page 73)


```python
# Y is  income
Y = np.linspace(0, 150, num=2) 

# starting values for c0, c1, G, I, and T
c10 = 10
c11 = 0.9
G = 10
I = 5
T = 3

# the demand function
def demand(c0, c1, Y, G, I, T):
    return (c0 + I + G - c1*T) + c1*Y

# equilibrium equation: production equals demand (demand equals income)
def equilibrium(Y):
    return Y

fig1, ax1 = plt.subplots()

# Draw the initial plot
surf11 = ax1.plot(Y, equilibrium(np.ravel(Y)), label="equilibrium equation")
surf12 = ax1.plot(Y, demand(c10, c11, np.ravel(Y), G, I, T), label="demand")

ax1.set_xlim([0, 100])
ax1.set_ylim([0, 100])

# set axis labels
ax1.set_xlabel("income(Y)")
ax1.set_ylabel('demand(Z) and production(Y)')
    

# set legend 
plt.legend(loc="upper left")

# update plot and show slides
slider_c10 = ipywidgets.IntSlider(value=10, min=0, max=100, description = "c0", continuous_update = True)
slider_c11 = ipywidgets.FloatSlider(value=0.5, min=0, max=1, description = "c1", continuous_update = True)

slider_G = ipywidgets.IntSlider(value=10, min=0, max=100, description = "G", continuous_update = True)
slider_I = ipywidgets.IntSlider(value=5, min=0, max=100, description = "I", continuous_update = True)

slider_T = ipywidgets.IntSlider(value=10, min=0, max=100, description = "T", continuous_update = True)

def update(c10=10, c11=0.9, G=10, I=5, T=3):
    # clearing and rebuilding the graph with new values for c0 and c1.
    # Not the most efficient method, but it gets the job done.
    ax1.clear()
    ax1.set_xlim([0, 100])
    ax1.set_ylim([0, 100])
    ax1.set_xlabel("income(Y)")
    ax1.set_ylabel('demand(Z) and production(Y)')
    surf11 = ax1.plot(Y, equilibrium(np.ravel(Y)), label="equilibrium equation")
    surf12 = ax1.plot(Y, demand(c10, c11, np.ravel(Y), G, I, T), label="demand")
    ax1.legend(loc="upper left")
    
    
   

interactive_plot_2 = ipywidgets.interactive(update, c10=slider_c10, c11=slider_c11, G=slider_G, I=slider_I, T=slider_T)
display(interactive_plot_2)
```
